```python
fig = plt.figure()
fig.suptitle('My SubPlots')
fig.add_subplot(221)
plt.plot([np.log(n) for n in range(1,10)])
fig.add_subplot(222, facecolor='y')
fig.add_subplot(223)
plt.plot(np.random.randn(10))
fig.add_subplot(224, facecolor='c')
plt.plot(np.random.rand(10), color ='k')
plt.show()
```

```python
fig = plt.figure()
fig.suptitle('My SubPlots')
fig.add_subplot(331)
plt.plot([np.log(n) for n in range(1,10)])
fig.add_subplot(332, facecolor='y')
fig.add_subplot(333)
plt.plot(np.random.randn(10))
fig.add_subplot(334)
plt.plot(np.random.random(10))
fig.add_subplot(335)
plt.plot(np.random.rand(10))
fig.add_subplot(336)
plt.plot(np.random.randn(100))
plt.show()
```

```python
#plots stacked on top of another, sharing an axis
fig, plots = plt.subplots(2, sharex=True)
fig.suptitle("Sharing X axis")
x = range(0, 1000, 25)
y = [n**0.8 for n in x]
plots[0].plot(x, y, color='c')
plots[1].scatter(x, y, color = 'r')
plt.show()
```
