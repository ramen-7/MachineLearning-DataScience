### Normilzation using MinMaxScaler

```python
def MinMaxScaling(df, X):
    max = df[X].max(axis=0)
    min = df[X].min(axis=0)
    X_std = (df[X] - min / max - min)
    df[X] = X_std * (max - min) + min
```

```python
MinMaxScaling(df, 'YearlyIncome')
```

```python
df.YearlyIncome
```

### Label Encoding

```python
def LabelEncoding(df, X):
    unique_vals = df[X].unique()
    hashmap = {}
    for i in range(0, len(unique_vals)):
        hashmap[unique_vals[i]] = i
    df[X] = df[X].map(hashmap)
```

```python
LabelEncoding(df, 'Education')
```

```python
LabelEncoding(df, 'Occupation')
```

```python
LabelEncoding(df, 'City')
```

```python
LabelEncoding(df, 'StateProvinceName')
```

```python
LabelEncoding(df, 'CountryRegionName')
```

```python
LabelEncoding(df, 'PostalCode')
```

#### One Hot Encoding
`Gender` and `MaritalStatus` are to be One Hot Encoded

```python
for i in range(0, len(df)):
    if df['Gender'][i] == 'M':
        df['Gender'][i] = 1
    else:
        df['Gender'][i] = 0
```

```python
maritalstatus=[]
for i in range(0, len(df)):
    if df['MaritalStatus'][i] == 'M':
        maritalstatus.append(1)
    else:
        maritalstatus.append(0)
print(len(maritalstatus))
```

```python
df.insert(7, "MaritalState", maritalstatus)
```
