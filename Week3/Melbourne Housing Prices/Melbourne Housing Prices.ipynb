```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
df = pd.read_csv('Melbourne_housing_FULL.csv')
pd.set_option('display.max_columns', None)
```

```python
df.drop(['Address', 'Date', 'Lattitude', 'Longtitude'], axis=1, inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
na_fill_cols = ['Propertycount', 'Bathroom', 'Bedroom2', 'Car', 'Distance', 'YearBuilt']
df[na_fill_cols] = df[na_fill_cols].fillna(0)
```

```python
df.isna().sum()
```

```python
mean_landsize = df.Landsize.mean()
mean_building_area = df.BuildingArea.mean()
#mean_price = df.Price.mean()
```

```python
df.Landsize = df.Landsize.fillna(mean_landsize)
df.BuildingArea = df.BuildingArea.fillna(mean_building_area)
#df.Price = df.Price.fillna(mean_price)
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df = pd.get_dummies(df, drop_first=True)
df
```

```python
X = df.drop(['Price'], axis=1)
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 0.3, random_state = 7)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(train_X, train_y)
```

```python
model.score(test_X, test_y)
```

```python
model.score(train_X, train_y)
```

```python
from sklearn import linear_model

lasso_model = linear_model.Lasso(alpha=50, max_iter=100, tol=0.17)

lasso_model.fit(train_X, train_y)
```

```python
lasso_model.score(train_X, train_y)
```

```python
lasso_model.score(test_X, test_y)
```

```python
from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha=50, max_iter=100, tol=0.1)
```

```python
ridge_model.fit(train_X, train_y)
```

```python
ridge_model.score(test_X, test_y)
```

```python
ridge_model.score(train_X, train_y)
```
